```python
import pandas as pd
import os
import cv2 as cv
import random
from skimage.transform import resize
from skimage.io import imread
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
dir = r'C:\Users\Bhavana\OneDrive\Documents\PDA\Machine_Learning\Internship-PRODIGY\Dataset'
```

```python
categories = ['Cat', 'Dog']

data = []

for category in categories:
    path = os.path.join(dir, category)
    label = categories.index(category)

    for img in os.listdir(path):
        imgpath = os.path.join(path, img)
        pet_img = cv.imread(imgpath, 0)
        try:
            pet_img = cv.resize(pet_img, (100, 100), interpolation=cv.INTER_AREA)
        except:
            break
        image = np.array(pet_img).flatten()
        data.append([image,label])

print(len(data))
```

```python
random.shuffle(data)

features = []
labels = []

for feature,label in data:
    features.append(feature)
    labels.append(label)
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.2)
```

```python
model = SVC(C=1, kernel='poly', gamma='auto') 
model.fit(x_train,y_train)
```

```python
mypet = x_test[1].reshape(100,100)
prediction = model.predict(x_test)
accuracy = model.score(x_test,y_test)
print('Accuracy :', accuracy)
print('Prediction is', categories[prediction[2]])
plt.imshow(mypet,cmap='gray')
plt.show()
```
